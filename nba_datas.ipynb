```python
from basketball_reference_web_scraper import client
from basketball_reference_web_scraper.data import OutputType
from nba_api.stats.endpoints import playercareerstats
import json
import pandas as pd

PLAYER_JSON = client.players_advanced_season_totals(season_end_year=2023, output_type=OutputType.JSON)

total_t = json.loads(PLAYER_JSON)
PLAYER_LIST = {}

for d in total_t:
    for k, v in d.items():
        if k in PLAYER_LIST :
            if isinstance(PLAYER_LIST[k], list):
                PLAYER_LIST[k].append(v)
            else:
                PLAYER_LIST[k] = [PLAYER_LIST[k], v]
        else:
            PLAYER_LIST[k] = v

for k, v in PLAYER_LIST.items():
    if isinstance(v, list) and len(v) == 1:
        PLAYER_LIST[k] = v[0]

print(PLAYER_LIST.keys())

PLAYER_stats = pd.DataFrame(PLAYER_LIST)

display(PLAYER_stats)

```
