```python
from basketball_reference_web_scraper import client
from basketball_reference_web_scraper.data import OutputType
from nba_api.stats.endpoints import playercareerstats
import json
import pandas as pd
```

```python
potential_player = {}

PLAYER_JSON = client.players_season_totals(season_end_year=2021, output_type=OutputType.JSON)
total_t = json.loads(PLAYER_JSON)

PLAYER_stats = pd.json_normalize(total_t)

# get the players who play more than 30 minutes

Player_list_1 = PLAYER_stats[PLAYER_stats['minutes_played']/PLAYER_stats['games_played'] >= 30]

new_columns = ['name'] + ['minutes_played'] + ['games_played'] + ['points'] + ['team'] + [c for c in Player_list_1.columns if c not in ('name','minutes_played','games_played','points','team')]
Player_list_1 = Player_list_1[new_columns]

#print(Player_list_1.columns)
Player_list_1 = Player_list_1.sort_values(by='points', ascending=False)
display(Player_list_1)

```

```python
PLAYER_advanced_JSON = client.players_advanced_season_totals(season_end_year=2023, output_type=OutputType.JSON)
advanced_total_t = json.loads(PLAYER_advanced_JSON)

PLAYER_advanced_stats = pd.json_normalize(advanced_total_t)
advanced_new_column = ['name'] + [c for c in PLAYER_advanced_stats.columns if c != 'name']
PLAYER_advanced_stats = PLAYER_advanced_stats[advanced_new_column]
PLAYER_advanced_stats = PLAYER_advanced_stats.sort_values(by='player_efficiency_rating', ascending=False)
print(PLAYER_advanced_stats.columns)

display(PLAYER_advanced_stats)
```
