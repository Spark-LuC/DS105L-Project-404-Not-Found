```python
import requests

import pandas as pd

from bs4 import BeautifulSoup
```

```python
my_url = 'https://www.bbc.co.uk/search?q=james+mvp&d=SPORT_PS'
response = requests.get(my_url)
soup = BeautifulSoup(response.text)

event_cards = soup.find_all('div', attrs={"data-testid": "default-promo", "class":"ssrcss-53phst-Promo ett16tt0"})
```

```python
base_url = 'https://www.bbc.co.uk/search'
player_list = ["LeBron+James", "Nikola+Jokić", "Giannis+Antetokounmpo", "James+Harden", "Russell+Westbrook", "Stephen+Curry", "Kevin+Durant", "Derrick+Rose", "Kobe+Bryant", "Dirk+Nowitzki", "Steve+Nash"]
full_result = []
number_result = []
for player in player_list:
    results = []
    page_number = 1

    while True:
        params = {'q': player + '+mvp', 'd': 'SPORT_PS', 'page': str(page_number)}

        try:
            response = requests.get(base_url, params=params)
            soup = BeautifulSoup(response.text, 'html.parser')
            event_cards = soup.find_all('div', attrs={"data-testid": "default-promo", "class":"ssrcss-53phst-Promo ett16tt0"})

            if not event_cards:
                break

            for card in event_cards:
                title = card.find('p', attrs={'class': "ssrcss-6arcww-PromoHeadline e1f5wbog5"}).text.strip()
                date = card.find('span', attrs={'class': 'ssrcss-1if1g9v-MetadataText ecn1o5v1'}).text.strip()
                url = card.find('a').get('href')
                results.append({'title': title, 'date': date, 'url': url})

            page_number += 1

        except Exception as e:
            break
        
    full_result.append(results)
    number_result.append({"player": player, "number of results": len(results)})
    
```
